```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import requests
import bs4

import re
```

```python
#HM - Hamburger
#CF - Chicken fillet
#FF - Fish fillet
daily_sales = os.path.join('data', 'daily_sales.csv')
daily = pd.read_csv(daily_sales)
daily.head()
```

```python
"""This is all prep just for the Day column to be used"""
#regex for getting just day in Day column
r = r"\/([0-9_]+)\/"

#turning object in Day column into Strings
str_day = []
for obj in daily["Day"]:
    str_day.append(obj)  

#using regex on each element in Day column to get just Day
just_day = []
for date in str_day:
    just_day.append(re.search(r, date).group(1))
    
#converting String form of just Day into Int form
int_day = []
for day in just_day:
    int_day.append(int(day))

#daily.plot(x="Day", y=["HM-NE"]).set_xticks(int_day)
#daily.plot(x="Day", y=["CF-NE", "CF-NW", "CF-SE", "CF-C"])
#daily.plot(x="Day", y=["FF-NE", "FF-NW", "FF-SE", "FF-C"])
```

```python
#Set X values as int form of just Day
x = int_day

#HM-NE column used as the Y values in Int form
HM_NE_sales = []
for i in daily["HM-NE"]:
    HM_NE_sales.append(int(i))
HM_NE_y = HM_NE_sales


# Calculate the simple average of the data
HM_NE_y_mean = [np.mean(HM_NE_y)]*len(x)

fig, ax = plt.subplots()

# Plot the data
HM_NE_sales_line = ax.plot(x,HM_NE_y, label='HM_NE_Sales', marker='o')

# Plot the average line
HM_NE_mean_line = ax.plot(x,HM_NE_y_mean, label='Mean', linestyle='--')

# Make a legend
legend = ax.legend(loc='lower left')

plt.show()
```
